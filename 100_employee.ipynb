```python
names = ["Zia","Fiza","Shiza" ,"Kanwal","Komal","Nida","Rida" ,"Ali" ,"Rooma" ,"Uzma"]
departments = ["javascript" ,"HTML" ,"Css","marketing","engineering","sales" ,"finence" ,"Ai generate"]
skills =["javascript" ,"HTML" ,"Css","c++","Next j.s","Ai" ,"data science" ,"Ai generate" ,"python"]

employees =[]
import random
for i in range(1 ,1001):
    emp = {
        "id" : i,
        "name" :names [i % len(names)],
        "age" : random.randint(20,60),
        "department":departments[i % len(departments)],
        "skill " : skills[i% len(skills)],
        "salary" : random.randint(50000,100000)
    }
    employees.append(emp)
for e in employees[:3]:
    print(e)
```
